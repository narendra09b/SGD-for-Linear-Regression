<h2> SGD Classifier for Linear Regression </h2>

<h5> Loading Data from SKLearn </h5>

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston.data.shape)
```

```python
print(boston.feature_names)
```

```python
print(boston.target)
```

```python
print(boston.DESCR)
```

```python
import pandas as pd
bos = pd.DataFrame(boston.data)
print(bos.head())
```

```python
bos['PRICE'] = boston.target

X = bos.drop('PRICE', axis = 1)
Y = bos['PRICE']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.linear_model import SGDRegressor
clf = SGDRegressor(loss='squared_loss')
clf.fit(X_train, Y_train)
```

```python
import matplotlib.pyplot as plt
Y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

train_mse = mean_squared_error(Y_train, clf.predict(X_train))
test_mse = mean_squared_error(Y_test, Y_pred)
print('Train Mean Square error: ', train_mse)
print('Test Mean Square error: ', test_mse)
```

```python
performance = []
performance.append(['Regressor', 'Training MSE', 'Testing MSE'])
performance.append(['SGDRegressor', round(train_mse,2), round(test_mse,2)])
```

```python
plt.scatter(Y_test, Y_pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
plt.show()
```

```python
delta_y = Y_test - Y_pred;

import seaborn as sns;
import numpy as np;
sns.set_style('whitegrid')
sns.kdeplot(np.array(delta_y), bw=0.5)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.kdeplot(np.array(Y_pred), bw=0.5)
plt.show()
```

<h2> SGD Implementation </h2>

```python
def sgd_predict(values, coordinates):
    res = 0
    for i in range(0, len(coordinates)-1):
        res += values[i] * coordinates[i]
    res += coordinates[len(coordinates)-1]
    return res
```

```python
def sgd_optimise(x_train, coordinates, y_train, 
                 learning_rate, n_ephochs):
    for i in range(n_ephochs):

        total_error = 0
        for j in range(len(x_train)):
            err = y_train.iloc[j] - sgd_predict(x_train[j], coordinates) 
            total_error += abs(err)
            
            for k in range(len(coordinates)-1):
                coordinates[k] = coordinates[k] \
                                + (learning_rate * err * x_train[j][k]) 
            
            coordinates[len(coordinates)-1] = coordinates[len(coordinates)-1] \
                                            + (learning_rate * err)
        if i % 100 == 0:
            print('Epoch: ', i, 'Error: ', round(total_error,3))
    return coordinates
```

```python
import numpy as np
coordinates = np.random.rand(len(X_train[0])+1)
coordinates = sgd_optimise(X_train, coordinates, Y_train, 0.0001, 101)
```

```python
train_pred = []
for row in X_train:
    train_pred.append(sgd_predict(row, coordinates))   
    
test_pred = []
for row in X_test:
    test_pred.append(sgd_predict(row, coordinates))
```

```python
from sklearn.metrics import mean_squared_error

train_mse = mean_squared_error(Y_train, train_pred)
test_mse = mean_squared_error(Y_test, test_pred)
print('Train Mean Square error: ', train_mse)
print('Test Mean Square error: ', test_mse)
```

```python
performance.append(['Function', round(train_mse,2), round(test_mse,2)])
```

```python
plt.scatter(Y_test, test_pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
plt.show()
```

```python
delta_y = Y_test - test_pred

import seaborn as sns;
import numpy as np;
sns.set_style('whitegrid')
sns.kdeplot(np.array(delta_y), bw=0.5)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.kdeplot(np.array(test_pred), bw=0.5)
plt.show()
```
